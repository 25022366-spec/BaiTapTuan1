```python
def _shh(a) :
    t = 0 
    for i in range(1,a):
        if a % i == 0 :
            t+= i
    if t == a :
        return "hoan hao"
    return "khong hoan hao"
assert _shh(6) == ("hoan hao")
assert _shh(11) == ("khong hoan hao")
assert _shh(2) == ("khong hoan hao")
assert _shh(9) == ('khong hoan hao')
assert _shh(7) == ('khong hoan hao')

print("all tests accepted")

```

```python
def _Calculator(a, op,b):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return round(a/b,1)

assert _Calculator(8,"*", 7 ) == (56)
assert _Calculator(8,"-", 7 ) == (1)
assert _Calculator(8,"+", 7 ) == (15)
assert _Calculator(8,"/", 7 ) == (1.1)
assert _Calculator(0,"*", 0 ) == (0)
print("all tests accepted")
```

### W5A19
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.
Một số được gọi là số kỳ diệu nếu tổng các chữ số của nó (thực hiện lặp lại cho đến khi còn 1 chữ số) kết thúc bằng 1.


```python
def _mg(n):
    t = n
    while t >= 10:
        s = 0
        while t > 0:
            s += t % 10 
            t //= 10        
        t = s          
    
    return t == 1
assert _mg(1234 ) == (True)
assert _mg(987) == (False)
assert _mg(111 ) == (False)
assert _mg(199) == (True)
assert _mg(36) == (False)
print("all tests accepted")

```
