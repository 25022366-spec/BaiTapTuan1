### W4A1 Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)

W4A13 Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
a, b = map(int, input().split())
s1 = 0
s2 = 0
for i in range(1, a ) :
    if a % i == 0 :
        s1 = s1 + i
for i in range(1, b ) :
    if b % i == 0 :
        s2 = s2 + i 
if s1 == b and s2 == a :
    print("True")
else :
    print("False")
```
